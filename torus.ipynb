```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
%matplotlib inline
```

```python
def line_function(inputs, equation,):
    return equation(inputs)


def mesh_function(template, line_):
    _, mesh = np.meshgrid(np.ones_like(template[0]), line_)
    return mesh

radius_equation=None
def _(var, min=2): return  np.sin(var)+min
radius_equation =_
radius_inputs = np.linspace(0, 2*np.pi, 200)
radius_line = line_function(radius_inputs, radius_equation)

theta_equation=None
def _(var): return var
theta_equation =_
theta_inputs = np.linspace(0+(np.pi/4), 2*np.pi+(np.pi/4), 200)
theta_line = line_function(theta_inputs, theta_equation)

T, R = np.meshgrid(theta_line, radius_line)

X = R * np.cos(T)
Y = R * np.sin(T)

Z_equation=None
def _(var): return np.sin(var)
Z_equation =_
Z_inputs =  np.linspace(0-(np.pi/2), 2*np.pi-(np.pi/2), len(X))
Z_line = line_function(Z_inputs, Z_equation)
Z = mesh_function(X, Z_line)
```

```python
fig = plt.figure(dpi=150)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface \
(X, Y, Z, cmap='winter', linewidth=1, 
antialiased=True, alpha=0.7, rstride=3, cstride=3)
# Labels.
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_zlim(-2, 2)

ax.elev = 50
ax.dist = 10
ax.azim = 148
```
